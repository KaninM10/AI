#Import

```python
import pandas as pd
import re
import nltk

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

#Download NLTK resources


```python
nltk.download('stopwords')
nltk.download('wordnet')

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

#Load Dataset

```python
data = pd.read_csv("/content/drive/MyDrive/AI/Week8/trum_tweet_sentiment_analysis.csv")

print(data.head())
print(data.columns)
```

Text Cleaning Function

```python
def text_cleaning_pipeline(text):
    # lowercase
    text = str(text).lower()

    # remove urls
    text = re.sub(r"http\S+|www\S+", "", text)

    # remove mentions, hashtags, punctuation, emojis
    text = re.sub(r"@[A-Za-z0-9_]+", "", text)
    text = re.sub(r"#", "", text)
    text = re.sub(r"[^\w\s]", "", text)

    # remove numbers
    text = re.sub(r"\d+", "", text)

    # tokenize
    tokens = text.split()

    # remove stopwords
    tokens = [w for w in tokens if w not in stop_words]

    # lemmatization
    tokens = [lemmatizer.lemmatize(w) for w in tokens]

    return " ".join(tokens)
```

Load

```python
X = data["text"]
y = data["Sentiment"]
```

clean

```python
data["cleaned_text"] = data["text"].apply(text_cleaning_pipeline)
```

Train-test split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    data["cleaned_text"],
    data["Sentiment"],
    test_size=0.2,
    random_state=42
)
```

TF-IDF Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()

X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

Train model

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train_tfidf, y_train)
```

Predict

#

```python
y_pred = model.predict(X_test_tfidf)
```

Evaluation

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
